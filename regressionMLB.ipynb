```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mlb_teams_classified.csv") # change source file route if needed
df
```

```python
my_list = list(df)
print(*my_list,sep="\n")
```

```python
# Remove columns
df2=df.drop(["Unnamed: 0",
             "year",
             "losses",
            "team_name",
            "games_played",
            "class"],
            axis=1)
df2
# df2.dtypes
```

```python
my_list2 = list(df2)
print(*my_list2,sep="\n")
```

```python
# Assigning independent and dependent variables
X=df2.drop("wins", axis=1)
y=df2['wins'].values
```

```python
# Function to get multiple regression model stats
import statsmodels.api as sm
def get_stats():
    results = sm.OLS(y, X).fit()
    print(results.summary())
    # p_values=results.pvalues #extracts p-values only
    # print(p_values)
```

```python
get_stats()
```

```python
# Start of stepwise regression, choosing which variables to drop based on p-value (one at a time)
X=X.drop("hits",axis=1)
get_stats()
```

```python
X=X.drop("double_plays",axis=1)
get_stats()
```

```python
X=X.drop("fielding_percentage",axis=1)
get_stats()
```

```python
X=X.drop("doubles",axis=1)
get_stats()
```

```python
X=X.drop("triples",axis=1)
get_stats()
```

```python
X=X.drop("hits_allowed",axis=1)
get_stats()
```

```python
X=X.drop("caught_stealing",axis=1)
get_stats()
```

```python
X=X.drop("sacrifice_flies",axis=1)
get_stats()
```

```python
X=X.drop("strikeouts_by_batters",axis=1)
get_stats()
```

```python
X=X.drop("batters_hit_by_pitch",axis=1)
get_stats()
```

```python
X=X.drop("stolen_bases",axis=1)
get_stats()
```

```python
X=X.drop("homeruns_allowed",axis=1)
get_stats()
```

```python
X=X.drop("homeruns",axis=1)
get_stats()
```

```python
# Independent variables after stepwise regression
print(X.columns)
print(len(X.columns)," variables")
```

```python
# Testing for collinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

```python
# From theresults above, we can see that "opponents_runs_scored" and "earned_runs_allowed" are highly correlated. Thus we will drop "earned_runs_allowed" from the dataset
X=X.drop("earned_runs_allowed",axis=1)
```

```python
# Independent variables after collinearity test
print(X.columns)
print(len(X.columns)," variables")
```

```python
# Create CSV file with final dataset
final_df=df.drop(["Unnamed: 0",
             "hits",
             "double_plays",
            "fielding_percentage",
            "doubles",
            "triples",
            "hits_allowed",
            "caught_stealing",
            "sacrifice_flies",
            "strikeouts_by_batters",
            "batters_hit_by_pitch",
            "stolen_bases",
            "homeruns_allowed",
            "homeruns",
            "earned_runs_allowed"
            ],
            axis=1)
final_df.to_csv("mlb_teams_final.csv")
final_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
score = model.score(X, y)
print(f"R2 Score: {score}")
print('y-axis intercept: ', model.intercept_) 
list(zip(X, model.coef_))
coeff_table=pd.DataFrame(list(zip(X, model.coef_)))
coeff_table = coeff_table.rename(columns={0: 'X', 1: 'Coefficient'})
coeff_table.to_csv("coefficient_table.csv")
coeff_table

```

```python
import seaborn as sns
sns.set()
cols = ['wins',
        'runs_scored',
        'walks',
        'opponents_runs_scored',
        'complete_games',
        'shutouts',
        'saves',
        'outs_pitches',
        'walks_allowed',
        'strikeouts_by_pitchers',
        'errors']
sns.pairplot(df2[cols], kind='reg', plot_kws={'line_kws':{'color':'red'}}, height = 2.5) 
plt.savefig('pairplots.png')
plt.show()
```

```python
predictions = model.predict(X)
predictions_table=pd.DataFrame({"Predicted": predictions, "Actual": y, "Error": predictions - y})[["Predicted", "Actual", "Error"]]
predictions_table.to_csv("predictions_table.csv")
predictions_table
```

```python
predictions = model.predict(X)
plt.scatter(predictions, predictions - y)
plt.hlines(y=0, xmin=predictions.min(), xmax=predictions.max())
plt.show()
```

```python
# Fit the model to the training data, and calculate the scores for the training and testing data.

model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
