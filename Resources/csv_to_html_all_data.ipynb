```python
# Import dependencies
import pandas as pd
```

```python
# Read mlb_teams.csv into a Pandas DataFrame
mlb_teams_df = pd.read_csv("mlb_teams.csv")
mlb_teams_df
```

```python
mlb_teams_df.columns
```

```python
mlb_teams_df = mlb_teams_df[['year', 'team_name', 'league_id', 'division_id', 'rank', 'games_played',
       'home_games', 'wins', 'losses', 'division_winner', 'wild_card_winner',
       'league_winner', 'world_series_winner', 'runs_scored', 'at_bats',
       'hits', 'doubles', 'triples', 'homeruns', 'walks',
       'strikeouts_by_batters', 'stolen_bases', 'caught_stealing',
       'batters_hit_by_pitch', 'sacrifice_flies', 'opponents_runs_scored',
       'earned_runs_allowed', 'earned_run_average', 'complete_games',
       'shutouts', 'saves', 'outs_pitches', 'hits_allowed', 'homeruns_allowed',
       'walks_allowed', 'strikeouts_by_pitchers', 'errors', 'double_plays',
       'fielding_percentage', 'ball_park', 'home_attendance']]
mlb_teams_df.head(18)
```

```python
# Convert the DataFrame into an html table
html = mlb_teams_df.to_html(index=False)

file = open("mlb_teams_all_data.html", "w")
file.write(html)
file.close()
```
